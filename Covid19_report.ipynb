    4.1 Xác định ngày đầu tiên bộ dữ liệu bắt đầu lưu trữ số ca Covid của 3 bộ dữ liệu

<a id='section_id_4.1'></a>

```python
#4.1.1: Kiểm tra bảng số lượng ca nhiễm (confirmed_df)

# Tạo cột Tính tổng số ca nhiễm lũy kế theo ngày:
confirmed_check = pd.DataFrame({'Lũy_kế' : confirmed_df.iloc[0:,4:].sum(axis=0)}).copy()
#Tạo cột ngày tháng theo format chuẩn
confirmed_check.insert(0, 'Ngày', pd.to_datetime(confirmed_check.index))
confirmed_check.set_index('Ngày', inplace=True)
print('Kiểm tra bảng dữ liệu "số ca nhiễm":\n',confirmed_check.sort_index(ascending=True).head(5))
# Ta thấy dữ liệu được lưu trữ ngay từ những ngày đầu tiên, do vậy ngày đầu tiên trong bảng dữ liệu là ngày bắt đầu lưu trữ
print('Ngày bắt đầu lưu trữ dữ liệu là ngày đầu tiên trong bảng dữ liệu: ', confirmed_check.index.min(), '\n')
```

```python
#4.1.2: Kiểm tra bảng số ca tử vong (deaths_df)

## death_check để kiểm tra số lượng ca tử vong
death_check = pd.DataFrame({'Lũy_kế' : deaths_df.iloc[0:,4:].sum(axis=0)}).copy()
death_check.insert(0, 'Ngày', pd.to_datetime(death_check.index))
death_check.set_index('Ngày', inplace=True)
print('Kiểm tra bảng dữ liệu "số ca tử vong":\n',death_check.sort_index(ascending=True).head(5))
# Ta thấy dữ liệu được lưu trữ ngay từ những ngày đầu tiên, do vậy ngày đầu tiên trong bảng dữ liệu là ngày bắt đầu lưu trữ
print('Ngày bắt đầu lưu trữ dữ liệu là ngày đầu tiên trong bảng dữ liệu: ', death_check.index.min(), '\n')
```

```python
#4.1.3: Kiểm tra bảng số ca khỏi bệnh (recoveries_df)

## confirmed_check để kiểm tra số lượng ca khỏi bệnh
recover_check = pd.DataFrame({'Lũy_kế' : recoveries_df.iloc[0:,4:].sum(axis=0)}).copy()
recover_check.insert(0, 'Ngày', pd.to_datetime(recover_check.index))
recover_check.set_index('Ngày', inplace=True)
print('Kiểm tra bảng dữ liệu "số ca khỏi bệnh":\n',recover_check.sort_index(ascending=True).head(5))
# Ta thấy dữ liệu được lưu trữ ngay từ những ngày đầu tiên, do vậy ngày đầu tiên trong bảng dữ liệu là ngày bắt đầu lưu trữ
print('Ngày bắt đầu lưu trữ dữ liệu là ngày đầu tiên trong bảng dữ liệu: ', recover_check.index.min(), '\n')
```

```python
#4.2.1: Kiểm tra bảng số lượng ca nhiễm (confirmed_df)

## Ta tiếp tục kiểm tra bảng số lượng ca nhiễm, lọc ra những ngày có lượng ca nhiễm mới nhỏ hơn hoặc bằng 0
print(confirmed_check[confirmed_check['Lũy_kế'] == 0])

## Do không có ngày nào số lũy kế bằng 0 nên ta kết luận dữ liệu vẫn đang được cập nhật hàng ngày
print('Ngày cập nhật dữ liệu ca nhiễm gần nhất" : ', confirmed_check.index.max())
```

```python
# Điền code vào đây

#1. Nhóm số liệu mắc mới theo từng quốc gia hoặc khu vực
confirmed_gr = confirmed_df.groupby('Country/Region').agg('sum')
#2. Tạo bảng số ca mắc mới mỗi ngày
confirmed_daily = confirmed_gr.copy().drop(columns=['Lat', 'Long'], axis = 0).transpose()
#3. Tạo danh sách quốc gia để thống kê
country_lst = confirmed_daily.columns.values.tolist()
#4. Dùng hàm diff() để chuyển từ số liệu lũy kế sang số liệu hàng ngày
for x in country_lst: confirmed_daily[x] = confirmed_daily[x].diff()
confirmed_daily = confirmed_daily.transpose().drop(columns=['1/22/20'], axis = 0) # xóa ngày đầu tiên vì sau khi diff() sẽ có giá trị NaN
#5. Tạo danh sách ngày tháng
day = confirmed_daily.columns.values.tolist()
#6. Tạo cột số ca mắc mới nhiều nhất
confirmed_daily['max'] = confirmed_daily.max(axis=1)
#7. Tìm tên quốc gia có số ca mắc mới mỗi ngày nhiều nhất và in nó ra
QG = confirmed_daily['max'].idxmax()
print('Quốc gia có số người nhiễm một ngày lớn nhất là: ', QG)
#8. Tìm số ca mắc mới mỗi ngày nhiều nhất và in nó ra
SL = confirmed_daily['max'].max()
print('Số người nhiễm một ngày lớn nhất: ', int(SL))
#9. Tìm ngày có số ca mắc mới mỗi ngày nhiều nhất và in nó ra
for x in day:
    if confirmed_daily.loc[QG,x] == SL : 
        print('Vào ngày: ', str(pd.Timestamp(x))[0:10])
```

```python
def plot_params(ax,axis_label= None, plt_title = None,label_size=15, axis_fsize = 15, title_fsize = 20, scale = 'linear' ):
    # Tick-Parameters
    ax.xaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.yaxis.set_minor_locator(ticker.AutoMinorLocator())
    ax.tick_params(which='both', width=1,labelsize=label_size)
    ax.tick_params(which='major', length=6)
    ax.tick_params(which='minor', length=3, color='0.8')
    
    # Grid
    plt.grid(lw = 1, ls = '-', c = "0.7", which = 'major')
    plt.grid(lw = 1, ls = '-', c = "0.9", which = 'minor')

    # Vẽ Title
    plt.title( plt_title,{'fontsize':title_fsize})
    
    # Trục y
    plt.yscale(scale)
    plt.minorticks_on()
    # Vẽ các nhãn theo trục
    xl = plt.xlabel(axis_label[0],fontsize = axis_fsize)
    yl = plt.ylabel(axis_label[1],fontsize = axis_fsize)
```

```python
def visualize_covid_cases_5(confirmed, deaths,
                          continent=None , country = None , state = None, period = None, figure = None,
                          scale = "linear"):
    x = 0
    if figure == None:
        f = plt.figure(figsize=(10,10))
        # Sub plot
        ax = f.add_subplot(111)
    else :
        f = figure[0]
        # Sub plot
        ax = f.add_subplot(figure[1],figure[2],figure[3])
    ax.set_axisbelow(True)
    plt.tight_layout(pad=10, w_pad=5, h_pad=5)
    
    stats = [confirmed, deaths]
    label = ["Confirmed", "Deaths"]
    if continent != None:
        params = ["continent",continent]
    elif country != None:
        params = ["country",country]
    else: 
        params = ["All", "All"]
    color = ["blue","red","green"]
    marker_style = dict(linewidth=3, linestyle='-', marker='o',markersize=4, markerfacecolor='#ffffff')
    for i,stat in enumerate(stats):
        if params[1] == "All" :
            cases = np.sum(np.asarray(stat.iloc[:,5:]),axis = 0)[x:]
        else:
            cases = np.sum(np.asarray(stat[stat[params[0]] == params[1]].iloc[:,5:]),axis = 0)[x:]
        date = np.arange(1,cases.shape[0]+1)[x:]
        plt.plot(date,cases,label = label[i]+" (Total : "+str(cases[-1])+")",color=color[i],**marker_style)
        plt.fill_between(date,cases,color=color[i],alpha=0.3)
    if params[1] == "All" :
        Total_confirmed = np.sum(np.asarray(stats[0].iloc[:,5:]),axis = 0)[x:]
        Total_deaths = np.sum(np.asarray(stats[1].iloc[:,5:]),axis = 0)[x:]
    else :
        Total_confirmed =  np.sum(np.asarray(stats[0][stats[0][params[0]] == params[1]].iloc[:,5:]),axis = 0)[x:]
        Total_deaths = np.sum(np.asarray(stats[1][stats[1][params[0]] == params[1]].iloc[:,5:]),axis = 0)[x:]
        
    text = stats[0].columns[5]+" to "+stats[0].columns[-1]+"\n"
    text += "Mortality rate : "+ str(int(Total_deaths[-1]/(Total_confirmed[-1])*10000)/100)+"\n"
    text += "Last 5 Days:\n"
    text += "Confirmed : " + str(Total_confirmed[-1] - Total_confirmed[-6])+"\n"
    text += "Deaths : " + str(Total_deaths[-1] - Total_deaths[-6])+"\n"
    text += "Last 24 Hours:\n"
    text += "Confirmed : " + str(Total_confirmed[-1] - Total_confirmed[-2])+"\n"
    text += "Deaths : " + str(Total_deaths[-1] - Total_deaths[-2])+"\n"
    
    plt.text(0.02, 0.78, text, fontsize=15, horizontalalignment='left', verticalalignment='top', transform=ax.transAxes,bbox=dict(facecolor='white', alpha=0.4))
    
    # Vẽ các nhãn trục
    axis_label = ["Days ("+confirmed_df.columns[5]+" - "+confirmed_df.columns[-1]+")","No of Cases"]
    
    # Vẽ các tham số
    plot_params(ax,axis_label,scale = scale)
    
    # Vẽ Title
    if params[1] == "All" :
        plt.title("COVID-19 Cases World",{'fontsize':25})
    else:   
        plt.title("COVID-19: "+params[1] ,{'fontsize':25})
        
    # Vị trí Legend 
    l = plt.legend(loc= "best",fontsize = 15)
    
    if figure == None:
        plt.show()
```
